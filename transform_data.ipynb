# Player Information

```python
# Import downloaded csv of player info
player_info = 'raw_data/basic_player_stat.csv'
player_info_df = pd.read_csv(player_info)
player_info_df.head()
```

```python
# Change player name format
# Grab first and last name in variables
first_name = player_info_df['Name'].str.split().str.get(1)
last_name = player_info_df['Name'].str.split().str.get(0)
# Add individual columns and drop full name
clean_info = player_info_df.drop(['Name'],axis=1)
clean_info['First Name'] = first_name
clean_info['Last Name'] = last_name
# Remove the comma
clean_info['Last Name'] = clean_info['Last Name'].str.replace(',','')
```

```python
# Remove unnecessary columns and drop retired players
df1 = clean_info.drop(['Years Played','Position','Number','Age','High School','High School Location','Player Id','Current Team'],axis=1)
df1.rename(columns={"Current Status":"Status"}, inplace=True)
df1 = df1[df1.Status !='Retired']
```

```python
# Combine First and Last into one column called Player
df1['Player'] = df1['First Name'] + ' ' + df1['Last Name']
df1.drop(['First Name','Last Name'],axis=1, inplace=True)
df1.head()
```

```python
df1.rename(columns={"Height (inches)":"Height", "Weight (lbs)":"Weight"}, inplace=True)
df1.head()
```

```python
# Rename to make headings lowercase and set index to players' name
df1.rename(columns={"Player":"player","Birth Place":"birth_place","Birthday":"birthday","College":"college","Status":"status",
                   "Experience":"experience","Height":"height","Weight":"weight"},
                     inplace=True)
df1.set_index(['player'],inplace=True)
df1.head()
```
